```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
years = [1930, 1934, 1938, 1950, 1954, 1958, 1962, 1966, 1970, 1974, 1978, 1982, 1986, 1990, 1994, 1998, 2002, 2006, 2010, 2014, 2018, 2022]
```

```python
url = 'https://en.wikipedia.org/wiki/2014_FIFA_World_Cup'
response = requests.get(url)
content = response.text
```

```python
soup = BeautifulSoup(content, 'lxml')
```

```python
matches = soup.find_all('div', class_="footballbox")
matches
```

```python
for match in matches:
    print(match.find('th', class_='fhome').get_text())
    print(match.find('th', class_='fscore').get_text())
    print(match.find('th', class_='faway').get_text())
```

```python
# to create a table u have to create a list with the headings you want
home = []
score = []
away = []
for match in matches:
    home.append(match.find('th', class_='fhome').get_text())
    score.append(match.find('th', class_='fscore').get_text())
    away.append(match.find('th', class_='faway').get_text())
    
football_match = {'home': home, 'score': score, 'away': away}
football_matches = pd.DataFrame(football_match)
football_matches['year'] = '2014'
print(football_matches)
```

```python
def get_matches(year):
    url = f'https://en.wikipedia.org/wiki/{year}_FIFA_World_Cup'
    response = requests.get(url)
    content = response.text
    soup = BeautifulSoup(content, 'lxml')
    matches = soup.find_all('div', class_="footballbox")
    for match in matches:
        print(match.find('th', class_='fhome').get_text())
        print(match.find('th', class_='fscore').get_text())
        print(match.find('th', class_='faway').get_text())
    home = []
    score = []
    away = []
    for match in matches:
        home.append(match.find('th', class_='fhome').get_text())
        score.append(match.find('th', class_='fscore').get_text())
        away.append(match.find('th', class_='faway').get_text())
    
    football_match = {'home': home, 'score': score, 'away': away}
    football_matches = pd.DataFrame(football_match)
    football_matches['year'] = 'year'
    return(football_matches)
```

```python
get_matches('2022')
```

```python
games_in_fifa_worldcup = [get_matches(year) for year in years]
df_matches = pd.concat(games_in_fifa_worldcup, ignore_index = True)
df_matches.to_csv('fifa_world_matches.csv', index = False)
```
